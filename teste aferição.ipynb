### 1.b) Comum?: defina uma função que recebe duas listas como argumento e devolve o valor booleano verdadeiro se as listas tiverem pelo menos um elemento comum e falso caso contrário

```python

def comum(lista1,lista2):
    cond=False
    for elem in lista1:
        if elem in lista2:
            cond=True
    return cond

```

```python
# as listas já estão ordenadas 

def mergeListas(lista1,lista2):
    res=[]
    for elem in lista1:
        res.append(elem)
    for elem2 in lista2:
        res.append(elem2)
    res.sort()
    return res
    

```

### b) multiplos23: defina uma função que calcula quantos múltiplos de 23 existem no intervalo de 1 a n em que n é o argumento da função


```python
def multiplos23(n):
    res=0
    for num in range(1,n+1):
        if num%23==0:
            res=res+1
    return res


```

### d) reduxInt: defina uma função que recebe um número inteiro positivo como argumento e vai adicionar repetidamente os seu dígitos até obter apenas um dígito que é retornado como resultado

```python
def reduxInt(n):
    res=0
    final=0
    for c in str(n):
        res=res+int(c)
    for dig in str(res):
        final=final+int(dig)
    return final
    

print(reduxInt(998))
```

### d) existe_medico: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o nome
de um médico, e verifica se o médico existe nessa unidade hospitalar, devolvendo True caso exista e False em caso contrário.

```python
def existe_medico(hospital,nome_unidade,nome_med):
    existe=True
    for unid in hospital:
        if nome_unidade==unid[0]:
            if nome_med in unid[1][1]:
                existe=True
    return existe

print(existe_medico(hospital, "Unidade de Cardiologia", "Dra.Maria"))
```

### e) adicionar_medico: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o
nome de um médico, e caso este não exista na unidade, acrescenta-o e devolve o novo hospital com o
médico inserido

```python
def adicionar_medico(hospital,nome_unidade,nome_med):
    for unid in hospital:
        if nome_unidade==unid[0]:
            if nome_med not in unid[1][1]:
                unid[1].append(nome_med)
    return hospital

print(adicionar_medico(hospital, "Unidade de Cardiologia", "Dra.Paula"))
```

### f) ocupar_cama: Defina uma função que recebe um hospital, o nome de uma unidade hospitalar e o
identificador de uma cama, e caso a cama esteja livre devolve um novo hospital com a cama ocupada (não
faz nada se a cama estiver ocupada)

```python
def ocupar_cama(hospital,nome_unidade,id_cama):
    for unid in hospital:
        if nome_unidade==unid[0]:
            for cama in unid[2]:
                if id_cama==cama[0]:
                    if cama[1]==False:
                        unid[2].remove((id_cama,cama[1]))
                        unid[2].append((id_cama,True))               
    return hospital
print(ocupar_cama(hospital, "Unidade de Cardiologia", 101))
```
